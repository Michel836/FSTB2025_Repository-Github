```python
import random  # On importe la bibliothÃ¨que random pour gÃ©nÃ©rer un nombre alÃ©atoire

# GÃ©nÃ¨re un nombre alÃ©atoire entre 1 et 100 (inclus)
nombre_secret = random.randint(1, 100)

# On fixe le nombre maximum d'essais possibles
tentatives = 7

# On affiche une introduction pour l'utilisateur
print("Je pense Ã  un nombre entre 1 et 100. Tu as 7 essais.")

# On dÃ©marre la boucle de jeu : un tour par essai
for essai in range(1, tentatives + 1):
    # Demande Ã  l'utilisateur d'entrer un nombre pour cet essai
    essai_utilisateur = int(input(f"Essai {essai} : "))
    
    # Si l'utilisateur a trouvÃ© le bon nombre
    if essai_utilisateur == nombre_secret:
        print("Bravo, tu as trouvÃ© le nombre !")  # Message de victoire
        break  # On arrÃªte la boucle, car le jeu est gagnÃ©

    # Si le nombre proposÃ© est trop petit
    elif essai_utilisateur < nombre_secret:
        print("Trop bas ğŸ“‰")
    
    # Sinon, le nombre proposÃ© est trop grand
    else:
        print("Trop haut ğŸ“ˆ")

# Ce bloc else s'exÃ©cute seulement si la boucle for se termine sans break (donc sans victoire)
else:
    print(f"DÃ©solÃ©, tu as perdu. Le nombre Ã©tait : {nombre_secret}")
```
